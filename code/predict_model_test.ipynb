```python
# A, C 모델 로딩
from keras.models import load_model
model_A = load_model('/gdrive/MyDrive/Audio/model/ganada_A_220906.h5') 
model_C = load_model('/gdrive/MyDrive/Audio/model/ganada_C_220906.h5') 
```

# 앱에서 단계, 레이블, 녹음 파일(wav) 전달받으면
 - 레이블 index 결정
 - 단계에 해당하는 모델 사용

```python
test_label_0 = 0 # A 단계 label = 0 (아이) (전달받은 레이블 index)
test_label_1 = 1 # A 단계 1 (화가)
file_path = '/gdrive/MyDrive/Audio/test/10_test_data_30.wav' # 화가( 테스트용 wav) 전달받은 파일 path

import librosa
SAMPLING_RATE = 22050

test_x = []
test_data, _ = librosa.load(file_path, sr=SAMPLING_RATE)

test_x.append(test_data)
```

```python
# 테스트 코드 (들어보기용. 실제 실행시는 제거)
import IPython
print(test_label_1)
IPython.display.Audio(test_x[0], rate=SAMPLING_RATE)
```

```python
# padding
from tensorflow.keras.preprocessing import sequence

padded_test_x = sequence.pad_sequences(test_x, maxlen=22050, padding='post', dtype=float)
len(padded_test_x[0]) # 테스트용 코드
```

```python
# numpy array로 변경
import numpy as np
test_x = np.array(padded_test_x)
print(test_x.shape)  # 테스트용 코드
```

```python
# 테스트 : 레이블 확인 및 소리 듣기 
print(test_label_1)
IPython.display.Audio(test_x[0], rate=SAMPLING_RATE)
```

```python
y_ = model_A.predict(test_x)
```

```python
y_  # 각 레이블별 정확도 확인
```

```python
predicted = np.argmax(y_, axis=-1)
```

```python
# 정확도가 가장 높은 레이블의 정확도 확인
acc = y_[0]
acc[predicted[0]]  # 정확도
```

```python
# 모델 A로 '화가' 확인 : 100% 정확도
```

```python
# 정확도가 가장 높은 레이블의 id 확인
result_y = predicted[0]
result_y  # 레이블 ID 
```
