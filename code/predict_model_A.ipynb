# 앱에서 단계, 레이블, 녹음 파일(wav) 전달받으면
 - 레이블 index 결정
 - 단계에 해당하는 모델 사용

```python
test_label_0 = 0 # A 단계 label = 0 (아이) (전달받은 레이블 index)
file_path = '/gdrive/MyDrive/Audio/test/01_test_data_01.wav' # 전달받은 파일 path

import librosa
SAMPLING_RATE = 22050

test_x = []
test_data, _ = librosa.load(file_path, sr=SAMPLING_RATE)
test_x.append(test_data)
```

```python
# 테스트 코드 (들어보기용. 실제 실행시는 제거)
import IPython
print(test_label_0)
IPython.display.Audio(test_x[0], rate=SAMPLING_RATE)
```

```python
# 테스트용 코드
# len(test_data)
```

```python
# padding
from tensorflow.keras.preprocessing import sequence

padded_test_x = sequence.pad_sequences(test_x, maxlen=22050, padding='post', dtype=float)
len(padded_test_x[0]) # 테스트용 코드
```

```python
# numpy array로 변경
import numpy as np
test_x = np.array(padded_test_x)
print(test_x.shape)  # 테스트용 코드
```

```python
# 테스트용 코드
print(test_label_0)
IPython.display.Audio(test_x[0], rate=SAMPLING_RATE)
```

```python
# test_x[0]
```

```python
y_ = model_A.predict(test_x)  # 최초 실행시 11초, 이후 실행 시 1초 미만 (미리 실행해 두어 시간 단축 필요)
predicted = np.argmax(y_, axis=-1)

result_y = predicted[0]
result = False
if result_y == test_label_0 : # 레이블 0 : 아이
    result = True

# 결과 return 하여 앱에서 표시하도록 함
result 
```
